```python
# Removing duplicates 
sales_data = sales_data.drop_duplicates(keep = 'first')
```

```python
freq_publisher = sales_data['Publisher'].mode()[0]
```

```python
sales_data['Publisher'].fillna(freq_publisher, inplace = True)
```

```python
sales_data.info()
```

```python
sales_data['Region'].mode()[0]
```

```python
sales_data['Region'].fillna(sales_data['Region'].mode()[0], inplace = True)
```

```python
check = sales_data[sales_data['Publisher'].isna()]
check.shape[0]
```

```python
check2 = sales_data[sales_data['Region'].isna()]
check2.shape[0]
```

```python
sales_data['NA_Sales'] = sales_data['NA_Sales'].replace('[$]','',regex = True)
```

```python
sales_data.head(50)
```

```python
sales_data = sales_data.astype({'NA_Sales' : float})
```

```python
sales_data.dtypes
```

```python
sales_data = sales_data.rename(columns = {'NA_Sales':'National Sales [$]', 'Global_Sales':'Global Sales [$]', 'NA_Profit':'National Profit [$]', 'Global_Profit':'Global Profit [$]'})
```

```python
sales_data.head()
```

```python
sales_data.dtypes
```

```python
sales_data['Country'] = sales_data['Country'].replace('USA', 'United States')
```

```python
sales_data.head(50)
```

```python
# To make let's say the first letter of a name or a country uppercase
sales_data['Country'] = sales_data['Country'].str.title()
```

```python
sales_cap = sales_data['National Sales [$]'].quantile(0.99)
sales_cap
```

```python
sales_cap1 = sales_data['Global Sales [$]'].quantile(0.99)
sales_cap1
```

```python
sales_cap2 = sales_data['National Profit [$]'].quantile(0.99)
sales_cap2
```

```python
sales_cap3 = sales_data['Global Profit [$]'].quantile(0.99)
sales_cap3
```

```python
sales_data['National Sales [$]'] = np.where(sales_data['National Sales [$]'] > sales_cap, sales_cap, sales_data['National Sales [$]'])
sales_data['Global Sales [$]'] = np.where(sales_data['Global Sales [$]'] > sales_cap1, sales_cap1, sales_data['Global Sales [$]'])
sales_data['National Profit [$]'] = np.where(sales_data['National Profit [$]'] > sales_cap2, sales_cap2, sales_data['National Profit [$]'])
sales_data['Global Profit [$]'] = np.where(sales_data['Global Profit [$]'] > sales_cap3, sales_cap3, sales_data['Global Profit [$]'])
```

```python
sales_data.head(10)
```

```python
sales_data = sales_data.round({'National Sales [$]': 2, 'Global Sales [$]': 2, 'National Profit [$]':2, 'Global Profit [$]':2})
```

```python
sales_data.head(3)
```

```python
# Data Visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
National_Sales = sales_data.groupby(['Region','Country'])['National Sales [$]'].sum().reset_index().sort_values(by = 'National Sales [$]', ascending = False)
National_Sales
```

```python
# Create a bar chart
plt.figure(figsize = (16,8))
sns. barplot(data = National_Sales, x = 'Region', y = 'National Sales [$]', hue = 'Country')
plt.title('National Sales by Region and Country', fontweight = 'bold')
plt.xlabel('Region', fontweight = 'bold')
plt.ylabel('National Sales [$]', fontweight = 'bold')
plt.show()
```

```python
#Create a box plot
plt.figure(figsize = (12,6))
sns.boxplot(x = 'Country', y = 'National Sales [$]', data = sales_data, showmeans= True, hue = 'Genre', meanprops = {'marker': 'o', 
                                                                                                                     'markerfacecolor':'white',
                                                                                                                     'markersize': '6',
                                                                                                                     'markeredgecolor':'green'}
            )
plt.title('National Sales by Country', fontweight = 'bold')
plt.xlabel('Country', fontweight = 'bold')
plt.ylabel('National Sales [$]', fontweight = 'bold')
plt.show()
```

```python
#Create a pie chart
Sales = sales_data.groupby('Country')[['National Sales [$]','Global Sales [$]']].sum().reset_index()
Sales
```

```python
Country = Sales['Country']
National_Sales2 = Sales['National Sales [$]']
Global_Sales = Sales['Global Sales [$]']	
```

```python
fig, axs = plt.subplots(1,2,figsize = (10,7))
axs[0].pie(National_Sales2, labels= Country, autopct = '%1.1f%%', startangle = 90)
axs[0].set_title('National Sales by Country')

axs[1].pie(Global_Sales, labels = Country, autopct = '%1.1f%%', startangle = 90)
axs[1].set_title('Global Sales by Country')

plt.show()
```

```python
# Create a line chart
plt.figure(figsize = (10,6))

#Line for National Sales
plt.plot(sales_data['Year'], sales_data['National Sales [$]'], marker = 'o', linestyle = '-', color = 'b', label ='National Sales [$]')

#Line for Global Sales
#plt.plot(sales_data['Year'], sales_data['Global Sales [$]'], marker = 's', linestyle = '--', color = 'g', label ='Global Sales [$]')

#Adding titles and labels
plt.title('National and Global Sales Over Years', fontweight = 'bold')
plt.xlabel('Year', fontweight = 'bold')
plt.ylabel('Sales', fontweight = 'bold')
plt.legend()

#Display the line chart
plt.grid(True)
plt.show()
```
